```python
# Set the "release_date" column as the index of the 'track' DataFrame
track.set_index("release_date", inplace=True)

track.index=pd.to_datetime(track.index)
# Convert the index of the 'track' DataFrame to datetime format
track.head()
```

```python
track[["artists"]].iloc[18]

# Retrieve the value of the "artists" column at index 18
```

```python
# Calculate the duration in seconds by dividing 'duration_ms' by 1000 and rounding the result
track["duration"]=track["duration_ms"].apply(lambda x:round(x/1000))

# Remove the 'duration_ms' column from the 'track' DataFrame
track.drop("duration_ms", inplace=True, axis=1)

```

```python
#track['duration'].head()
track.duration.head()

# Display the first few rows of the 'duration' column in the 'track' DataFrame
```

```python
# Remove columns 'key', 'mode', and 'explicit' from the 'track' DataFrame
corr_df = track.drop(["key", "mode", "explicit"],axis=1).corr(method="pearson")
# Create a figure with a specific size for the heatmap plot
plt.figure(figsize=(14, 6))

# Generate a heatmap using seaborn to visualize the correlation matrix
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="black")

# Set the title for the correlation heatmap
heatmap.set_title("Correlation HeatMap between Variables")

# Rotate the x-axis labels for better readability
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)

```

```python
# Randomly sample a subset of the 'track' DataFrame
# The sample size is determined by multiplying 0.004 with the length of the DataFrame


sample_df=track.sample(int(0.004*len(track)))
```

```python
print(len(sample_df))

#The len() function returns the length of the DataFrame, 
```

```python
# Create a figure with a specific size for the regression plot
plt.figure(figsize=(10,6))

# Generate a regression plot using seaborn to visualize the relationship between 'loudness' and 'energy'
sns.regplot(data = sample_df,y="loudness",x="energy",color="c").set(title="Loudness vs Energy Correlation")
```

```python
# Create a figure with a specific size for the regression plot
plt.figure(figsize=(10,6))

# Generate a regression plot using seaborn to visualize the relationship between 'popularity' and 'acousticness'
sns.regplot(data = sample_df,y="popularity",x="acousticness",color="b").set(title="Popularity vs Acousticness Correlation")
```

```python
# Extract the 'release_date' values from the index and assign them to a new column 'dates'
track['dates']=track.index.get_level_values('release_date')

# Convert the 'dates' column to datetime format
track.dates=pd.to_datetime(track.dates)

# Extract the year component from the 'dates' column
years=track.dates.dt.year
```

```python
# Generate a histogram plot to visualize the number of songs per year
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="Number of songs per year")
```

```python
# Create a bar plot to visualize the average duration of songs per year

total_dr=track.duration
fig_dims =(18,7)
fig, ax=plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title="Year vs Duration")
plt.xticks(rotation=90)
```

```python
df_genre=pd.read_csv("C:\\Users\\HP\\Downloads\\DATA_ANALYSIS_PROJECT__1\\SpotifyFeatures.csv")
```

```python
# Set the title of the plot
plt.title("Duration of the songs in  Different Genres")
# Set the color palette for the plot
sns.color_palette("rocket",as_cmap= True)
# Create a bar plot to visualize the duration of songs in different genres
sns.barplot(y='genre',x="duration_ms",data=df_genre)
# Set the x-axis label
plt.xlabel("Duration in milli seconds")
# Set the y-axis label
plt.ylabel("Genres")
```

```python
# Set the style to darkgrid for the plot
sns.set_style(style="darkgrid")

# Set the figure size
plt.figure(figsize=(10, 5))

# Select the top 10 genres with the highest popularity
famous = df_genre.sort_values("popularity", ascending=False).head(10)

# Create a bar plot to visualize the popularity of the top 10 genres
sns.barplot(y="genre", x="popularity", data=famous).set(title="Top 5 Genres by Popularity")
```
